```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random as rand
```

```python
import matplotlib
matplotlib.use('TkAgg')
from pylab import *

global count
#n0 = 10 # number of agents
count = 0

dist_th = 100

class agent:
    pass

def distance(city1_x,city1_y,city2_x,city2_y):
    xDis = abs(city1_x-city2_x)
    yDis = abs(city1_y-city2_y)
    distance=np.sqrt((xDis**2)+(yDis**2))
    return distance




def initialize():
    print("initialize")
    global agents
    global agent_path
    global n
    n = randint(20)
    agents = []
    agent_path = []
    for i in range(n):
        ag = agent()
        ag.name = i
        ag.visited = 0
        ag.x = randint(1,n)
        ag.y = randint(1,n)
        ag.weight = randint(100)
        agents.append(ag)
    for i in range(n):
        ag = agent()
        ag.name = 0
        ag.x = 0
        ag.y = 0
        ag.distance = 0
        agent_path.append(ag)
        
def observe():
    print("observe")
    global agents
    global agent_path
    global n
    cla()
    nodes = []
    positions = []
    for i in range(len(agent_path)-1):
        node1 = agent_path[i].name
        node2 = agent_path[i+1].name
        d =  distance(agent_path[i].x,agent_path[i].y,agent_path[i+1].x,agent_path[i+1].y)
        nodes.append(np.r_[node1,node2,np.round(d)])
        #positions.append(np.r_[agent_path[i].x,agent_path[i].y])
        
    G = nx.Graph()
    for i in range(len(agents)):
        p = np.r_[agents[i].x,agents[i].y]
        G.add_node(agents[i].name,pos = p)
    G.add_weighted_edges_from(nodes)
    
    pos=nx.get_node_attributes(G,'pos')
    #nx.draw_networkx(G,pos=pos)
    labels = nx.get_edge_attributes(G,'weight')
    #pos = nx.spring_layout(G)
    #nx.draw_networkx_edge_labels(G,pos,with_labels=True,edge_labels=labels)
    nx.draw_networkx(G, pos=pos,node_color='g')
    nx.draw_networkx_edge_labels(G, pos=pos, edge_labels = labels)
    nx.draw_networkx_edges(G,pos,width=2,edge_color='r', arrows=False)
    plt.show()
    
def update():
    print("update")
    global agents
    global agent_path
    global n
    global count
    dist = []
    nodes = []
    total_dist = 0
    if(count > 4):
        n = n + 1
        agents = []
        agent_path = []
        for i in range(n):
            ag = agent()
            ag.name = i
            ag.visited = 0
            ag.x = randint(1,n)
            ag.y = randint(1,n)
            ag.weight = randint(100)
            agents.append(ag)
        for i in range(n):
            ag = agent()
            ag.name = 0
            ag.x = 0
            ag.y = 0
            ag.distance = 0
            agent_path.append(ag)
            
    if(total_dist < dist_th):
        source = agents[randint(n)]
        neighbors = [nb for nb in agents if(nb != source)]
        new_source = source
        flag= 0

        agent_path[flag].x = source.x
        agent_path[flag].y = source.y
        agent_path[flag].name = source.name

        while((np.sum([ag.visited for ag in agents]) < n) & (len(neighbors) != 0)):
            for nb in neighbors:
                d = distance(new_source.x,new_source.y,nb.x,nb.y)
                dist.append(d)
            nearest_agent = neighbors[np.argmin(dist)]
            if(nearest_agent.visited != 1):
                nearest_agent.visited = 1
                total_dist += nearest_agent.weight
                new_source = agents[np.argmin(dist)]
                ix = np.argmin(dist)
                del neighbors[ix]
                flag = flag +1
                agent_path[flag].x = nearest_agent.x
                agent_path[flag].y = nearest_agent.y
                agent_path[flag].name = nearest_agent.name
                dist = []

        d = distance(source.x,source.y,new_source.x,new_source.y)
        dist.append(d)
        if(source.visited != 1):
            source.visited = 1
            total_dist += source.weight
            agent_path[flag].x = source.x
            agent_path[flag].y = source.y
            agent_path[flag].name = source.name
        count = count +1
    
        
```

```python
import pycxsimulator
pycxsimulator.GUI().start(func=[initialize, observe, update])
```
