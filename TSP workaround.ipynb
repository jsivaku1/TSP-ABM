```python
def distance(city1_x,city1_y,city2_x,city2_y):
    xDis = abs(city1_x-city2_x)
    yDis = abs(city1_y-city2_y)
    distance=np.sqrt((xDis**2)+(yDis**2))
    return distance
```

```python
def initialize(n):
    global agents
    global agent_path
    agents = []
    agent_path = []
    for i in range(n):
        ag = agent()
        ag.name = i
        ag.visited = 0
        ag.x = random.randint(1,n)
        ag.y = random.randint(1,n)
        ag.weight = randint(100)
        agents.append(ag)
    for i in range(n):
        ag = agent()
        ag.name = 0
        ag.x = 0
        ag.y = 0
        agent_path.append(ag)
```

```python
n = 10 #number of agents
dist = []


initialize(n)
total_dist = 0
source = agents[randint(n)]
neighbors = [nb for nb in agents if(nb != source)]
new_source = source
flag= 0

agent_path[flag].x = source.x
agent_path[flag].y = source.y
agent_path[flag].name = source.name

while((np.sum([ag.visited for ag in agents]) < n) & (len(neighbors) != 0)):
    for nb in neighbors:
        d = distance(new_source.x,new_source.y,nb.x,nb.y)
        dist.append(d)
    nearest_agent = neighbors[np.argmin(dist)]
    if(nearest_agent.visited != 1):
        nearest_agent.visited = 1
        total_dist += nearest_agent.weight
        new_source = agents[np.argmin(dist)]
        ix = np.argmin(dist)
        del neighbors[ix]
        flag = flag +1
        agent_path[flag].x = nearest_agent.x
        agent_path[flag].y = nearest_agent.y
        agent_path[flag].name = nearest_agent.name
        dist = []

d = distance(source.x,source.y,new_source.x,new_source.y)
dist.append(d)
if(source.visited != 1):
    source.visited = 1
    total_dist += source.weight
```

```python
from sys import maxsize 
V = 4
  
# implementation of traveling Salesman Problem 
def travellingSalesmanProblem(graph, s): 
  
    # store all vertex apart from source vertex 
    vertex = [] 
    for i in range(V): 
        if i != s: 
            vertex.append(i) 
  
    # store minimum weight Hamiltonian Cycle 
    min_path = maxsize 
  
    while True: 
  
        # store current Path weight(cost) 
        current_pathweight = 0
  
        # compute current path weight 
        k = s 
        for i in range(len(vertex)): 
            current_pathweight += graph[k][vertex[i]] 
            k = vertex[i] 
        current_pathweight += graph[k][s] 
  
        # update minimum 
        min_path = min(min_path, current_pathweight) 
  
        if not next_permutation(vertex): 
            break
  
    return min_path 
  
# next_permutation implementation 
def next_permutation(L): 
  
    n = len(L) 
  
    i = n - 2
    while i >= 0 and L[i] >= L[i + 1]: 
        i -= 1
  
    if i == -1: 
        return False
  
    j = i + 1
    while j < n and L[j] > L[i]: 
        j += 1
    j -= 1
  
    L[i], L[j] = L[j], L[i] 
  
    left = i + 1
    right = n - 1
  
    while left < right: 
        L[left], L[right] = L[right], L[left] 
        left += 1
        right -= 1
  
    return True
  
# Driver Code 
if __name__ == "__main__": 
  
    # matrix representation of graph 
    graph = [[ 0., 10., 12.,  8.],
       [10.,  0., 18., 18.],
       [12., 18.,  0., 10.],
       [ 8., 18., 10.,  0.]]
    s = 0
    print(travellingSalesmanProblem(graph, s)) 
```
