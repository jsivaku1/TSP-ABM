```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random as rand
import pandas as pd
```

```python
import matplotlib
matplotlib.use('TkAgg')
from pylab import *

global count
#n0 = 10 # number of agents
count = 0

dist_th = 100

class agent:
    pass

def distance(city1_x,city1_y,city2_x,city2_y):
    xDis = abs(city1_x-city2_x)
    yDis = abs(city1_y-city2_y)
    distance=np.sqrt((xDis**2)+(yDis**2))
    return distance




def initialize():
    print("initialize")
    global agents
    global agent_path
    global city_agents
    global n_city
    global n_agent
    n_city = randint(15,20)
    n_agent = 5
    city_agents = []
    agents = []
    agent_path = []
    #city agent
    for i in range(n_city):
        ag = agent()
        ag.city_name = i
        ag.visited = np.zeros([n_agent])
        ag.x = randint(1,n_city)
        ag.y = randint(1,n_city)
        city_agents.append(ag)
        
    #multiple agents
    for i in range(n_agent):
        ag = agent()
        ag.agent_name = i
        agents.append(ag)
        
        #agent paths
    for i in range(n_agent):
        ag = agent()
        ag.agent_name = i
        ag.x = 0
        ag.y = 0
        ag.city_visited = []
        ag.city_to_not_visit = []
        agent_path.append(ag)
        
def observe():
    global agents
    cla()
    cityname = [ag.city_name for ag in city_agents]
    x = [ag.x for ag in city_agents]
    y = [ag.y for ag in city_agents]
    plt.scatter(x, y,marker='x', color='red')
    for i, txt in enumerate(cityname):
        plt.annotate('city'+str(txt), (x[i], y[i]))
        
    
    agentname = [ag.agent_name for ag in agent_path]
    x = [ag.x for ag in agent_path]
    y = [ag.y for ag in agent_path]
    plt.scatter(x, y,marker='o', color='green')
    for i, txt in enumerate(agentname):
        plt.annotate('agent'+str(txt), (x[i], y[i]))
        
        

    axis('image')
    
def nearest_agents():
    global agents
    global agent_path
    global city_agents
    global n_city
    global n_agent
    distMat = np.zeros((n_agent,n_agent))
    for i in range(len(agent_path)):
        for j in range(len(agent_path)):
            distMat[i,j] = distance(agent_path[i].x,agent_path[i].y,agent_path[j].x,agent_path[j].y)
    ix = np.argwhere(distMat == 0)[0]
    return ix[0],ix[1]
    
def update():
    print("update")
    global agents
    global agent_path
    global city_agents
    global n_city
    global n_agent
    global count
    dist = []
    nodes = []
    total_dist = 0
    source_lists = []
    city_list1 = []
    city_list2 = []
    for i in range(n_agent):
        source_lists.append(city_agents[randint(n_city)])
            
    
    for i in range(len(agents)):
        source = source_lists[i]

        neighbors = [nb_city for nb_city in city_agents if(nb_city != source)]
        print(len(neighbors))
        flag= 0

        agent_path[i].x=source.x
        agent_path[i].y=source.y
        agent_path[i].city_visited.append(source.city_name)
        source.visited[i] = 1

        #while((np.sum([np.sum(ag.visited) for ag in city_agents]) < n) & (len(neighbors) != 0)):
        for nb in neighbors:
            d = distance(source.x,source.y,nb.x,nb.y)
            dist.append(d)
        nearest_city = neighbors[np.argmin(dist)]
        min_dist = np.min(dist)
        if(nearest_city.city_name not in agent_path[flag].city_to_not_visit):
            if(nearest_city.visited[i] != 1):
                nearest_city.visited[i] = 1
                total_dist += distance(source.x,source.y,nearest_city.x,nearest_city.y)
                agent_path[i].agent_name = agents[i].agent_name
                agent_path[i].x=nearest_city.x
                agent_path[i].y=nearest_city.y
                agent_path[i].city_visited.append(nearest_city.city_name)
                source_lists[i] = city_agents[np.argmin(dist)]
                #ix = np.argmin(dist)
                #del neighbors[ix]
                flag = flag +1

        dist = []


        #swapping agents city to visit when they are on the same city, city visited by agent 1 
        #should not be visited by agent 2 now - an interaction
        ix1,ix2=nearest_agents()
        city_list1 = agent_path[ix1].city_visited
        city_list2 = agent_path[ix2].city_visited

        agent_path[ix1].city_to_not_visit = city_list2
        agent_path[ix2].city_to_not_visit = city_list1


        '''
        d = distance(source.x,source.y,new_source.x,new_source.y)
        dist.append(d)
        if(source.visited != 1):
            source.visited = 1
            total_dist += d
            agent_path[flag].x = source.x
            agent_path[flag].y = source.y
            agent_path[flag].agent_name = agents[flag].agent_name
        count = count +1
'''        
```

```python
import pycxsimulator
pycxsimulator.GUI().start(func=[initialize, observe, update])
```
